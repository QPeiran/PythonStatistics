```python
from scipy.stats import norm
# Plot the Maximum Likelihood Functions for different values of mu 
# and sigma
def plot_ll(x):
    plt.figure(figsize=(5,8))
    plt.title("Maximim Likelihood Functions")
    plt.xlabel("Mean Estimate")
    plt.ylabel("Log Likelihood")
    plt.ylim(-40, -30)
    plt.xlim(0, 12)
    mu_set = np.linspace(0, 16, 1000)
    sd_set = [.5, 1, 1.5, 2.5, 3, 3.5]
    max_val = max_val_location = None
    for i in sd_set:
        ll_array = []
        
        for j in mu_set:
            temp_mm = 0
            for k in x:
                temp_mm += np.log(norm.pdf(k, j, i)) # The LL function
            ll_array.append(temp_mm)
                
            if (max_val is None):
                max_val = max(ll_array)
            elif max(ll_array) > max_val:
                max_val = max(ll_array)
                max_val_location = j           
        # Plot the results
        plt.plot(mu_set, ll_array, label="sd: %.1f" % i)
        print ("The max LL for sd %.2f is %.2f" % (i, max(ll_array)))
       
    plt.axvline(x=max_val_location, color='black', ls='-.')
    plt.legend(loc='lower left')


x = [2, 3, 4, 5, 7, 8, 9, 10]
plot_ll(x)
```

```python
def myMLE(x):
    mu_set = np.linspace(0, 16, 1000)
    sd_set = [.5, 1, 1.5, 2.5, 3, 3.5]
    
    
    for i in sd_set:
        arr_mm = []
        for j in mu_set:
            temp_mm = 0
            for k in x:
                temp_mm += np.log(norm.pdf(k, j, i))
            arr_mm.append(temp_mm)
        print("The max LL for sd" , i , "is", max(arr_mm))

x = [2, 3, 4, 5, 7, 8, 9, 10]
myMLE(x)
```

```python
import matplotlib.animation as animation
from matplotlib.widgets import RadioButtons
from matplotlib.widgets import Slider

sample_size = 10000
normal = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
gamma = np.random.gamma(shape = 1.0, scale=1.0, size=sample_size)
uniform = np.random.uniform(low=0.0, high=10.0, size=sample_size)
exponential = np.random.exponential(scale=1.0, size=sample_size)

```

```python
fig, sub_plt = plt.subplots()
plt.subplots_adjust(top=0.65)


axcolor = 'lightgoldenrodyellow'
rax = plt.axes([0.05, 0.7, 0.25, 0.25], facecolor=axcolor)
radio = RadioButtons(rax, ('Normal', 'Gamma', 'Uniform', 'Exponential'))
```

```python
def dist_func(type_l):
    sub_plt.clear() # comment this line if you want to keep previous drawings
    dist_dict = {'Normal':normal, 'Gamma':gamma, 'Uniform':uniform, 'Exponential':exponential}
    data_type = dist_dict[type_l]
    hist = sub_plt.hist(data_type,bins=100)
    plt.draw()
    plt.show()

dist_func('Normal')
```
